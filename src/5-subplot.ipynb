### Exercise 5: Subplot Layout

```python
# TASK: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.
```

```python
import matplotlib.pyplot as plt
import numpy as np

def plot_2x2_subplots():
    try:
        # Data for the plots
        x = np.linspace(0, 10, 100)
        y_line = x ** 2
        categories = ['A', 'B', 'C', 'D']
        values = [5, 7, 3, 9]
        x_scatter = [1, 2, 3, 4, 5]
        y_scatter = [2, 4, 6, 8, 10]
        labels = ['Python', 'Java', 'C++', 'JavaScript']
        sizes = [40, 25, 20, 15]
        colors = ['skyblue', 'lightgreen', 'gold', 'salmon']

        # Create 2x2 subplot layout
        fig, axes = plt.subplots(2, 2, figsize=(12, 10))

        # 1. Line Plot
        axes[0, 0].plot(x, y_line, color='blue', linewidth=2, label='y = x²')
        axes[0, 0].set_title('Line Plot')
        axes[0, 0].set_xlabel('x values')
        axes[0, 0].set_ylabel('y values')
        axes[0, 0].grid(True)
        axes[0, 0].legend()

        # 2. Bar Chart
        axes[0, 1].bar(categories, values, color=['red', 'green', 'blue', 'orange'])
        axes[0, 1].set_title('Bar Chart')
        axes[0, 1].set_xlabel('Categories')
        axes[0, 1].set_ylabel('Values')
        axes[0, 1].grid(axis='y', linestyle='--', alpha=0.7)

        # 3. Scatter Plot
        axes[1, 0].scatter(x_scatter, y_scatter, color='purple', s=100, edgecolor='black', label='Data Points')
        axes[1, 0].set_title('Scatter Plot')
        axes[1, 0].set_xlabel('x values')
        axes[1, 0].set_ylabel('y values')
        axes[1, 0].grid(True)
        axes[1, 0].legend()

        # 4. Pie Chart
        axes[1, 1].pie(
            sizes,
            labels=labels,
            colors=colors,
            autopct='%1.1f%%',
            startangle=140,
            shadow=True
        )
        axes[1, 1].set_title('Pie Chart')
        axes[1, 1].axis('equal')  # Equal aspect ratio for a perfect circle

        # Adjust layout for better spacing
        plt.tight_layout()
        plt.show()

    except ImportError as e:
        print(f"Import error: {e}. Please ensure matplotlib and numpy are installed.")
    except Exception as e:
        print(f"An error occurred: {e}")

# Call the function
plot_2x2_subplots()
```
