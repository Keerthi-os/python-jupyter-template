### Exercise 3: Scatter Plot with Annotations

```python
# TASK: Create a scatter plot with x = [1, 2, 3, 4, 5] and y = [2, 4, 6, 8, 10].
# Annotate each point with its (x, y) value, and set the title as 'Scatter Plot with Annotations'.
```

```python
import matplotlib.pyplot as plt

def plot_scatter_with_annotations():
    try:
        # Data
        x = [1, 2, 3, 4, 5]
        y = [2, 4, 6, 8, 10]

        # Plot
        plt.figure(figsize=(8, 6))
        plt.scatter(x, y, color='purple', s=100, edgecolor='black', label='Data Points')

        # Annotate each point with (x, y) value
        for xi, yi in zip(x, y):
            plt.annotate(f'({xi}, {yi})', (xi, yi), textcoords="offset points", xytext=(5, 5), fontsize=10)

        # Customization
        plt.title('Scatter Plot with Annotations', fontsize=16)
        plt.xlabel('x values', fontsize=12)
        plt.ylabel('y values', fontsize=12)
        plt.grid(True, linestyle='--', alpha=0.7)
        plt.legend()

        # Display the plot
        plt.show()

    except ImportError as e:
        print(f"Import error: {e}. Please ensure matplotlib is installed.")
    except Exception as e:
        print(f"An error occurred: {e}")

# Call the function
plot_scatter_with_annotations()
```
