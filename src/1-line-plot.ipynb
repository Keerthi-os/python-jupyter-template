### Exercise 1: Line Plot Customization

```python
import matplotlib.pyplot as plt
import numpy as np

def plot_square_function():
    try:
        # Generate data
        x = np.linspace(0, 10, 100)
        y = x ** 2

        # Create the plot
        plt.figure(figsize=(10, 6))
        plt.plot(x, y, label='y = x²', color='blue', linewidth=2)

        # Customize the plot
        plt.title('Line Plot of y = x²', fontsize=16)
        plt.xlabel('x values', fontsize=12)
        plt.ylabel('y values', fontsize=12)
        plt.grid(True)
        plt.legend()

        # Display the plot
        plt.show()

    except ImportError as e:
        print(f"Import error: {e}. Make sure matplotlib and numpy are installed.")
    except Exception as e:
        print(f"An error occurred: {e}")

# Call the function
plot_square_function()
```
