### Exercise 2: Bar Plot with Colors

```python
# TASK: Create a bar plot with the following data: categories = ['A', 'B', 'C', 'D'] and values = [5, 7, 3, 9].
# Use different colors for each bar and add a title to the plot.
```

```python
import matplotlib.pyplot as plt

def plot_bar_chart():
    try:
        # Data
        categories = ['A', 'B', 'C', 'D']
        values = [5, 7, 3, 9]
        colors = ['red', 'green', 'blue', 'orange']  # Different colors for each bar

        # Plot
        plt.figure(figsize=(8, 6))
        plt.bar(categories, values, color=colors)

        # Customization
        plt.title('Bar Plot of Categories', fontsize=16)
        plt.xlabel('Categories', fontsize=12)
        plt.ylabel('Values', fontsize=12)
        plt.grid(axis='y', linestyle='--', alpha=0.7)  # Optional: Grid for better readability

        # Display the plot
        plt.show()

    except ImportError as e:
        print(f"Import error: {e}. Please ensure matplotlib is installed.")
    except Exception as e:
        print(f"An error occurred: {e}")

# Call the function
plot_bar_chart()
```
