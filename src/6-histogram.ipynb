### Exercise 6: Histogram Customization

```python
# TASK: Create a histogram for the following data: data = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4, 5].
# Customize the histogram with a title, labels for the x-axis, and a specific color for the bars.
```

```python
import matplotlib.pyplot as plt

def plot_histogram():
    try:
        # Data
        data = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4, 5]

        # Plot
        plt.figure(figsize=(8, 6))
        plt.hist(data, bins=5, color='teal', edgecolor='black')  # Custom color and bar borders

        # Customizations
        plt.title('Histogram of Data', fontsize=16)
        plt.xlabel('Value', fontsize=12)
        plt.ylabel('Frequency', fontsize=12)
        plt.grid(axis='y', linestyle='--', alpha=0.7)  # Add grid lines for better readability

        # Display the plot
        plt.show()

    except ImportError as e:
        print(f"Import error: {e}. Please ensure matplotlib is installed.")
    except Exception as e:
        print(f"An error occurred: {e}")

# Call the function
plot_histogram()
```
