### Exercise 4: Pie Chart with Percentages

```python
# TASK: Create a pie chart with the following data: labels = ['Python', 'Java', 'C++', 'JavaScript'] and sizes = [40, 25, 20, 15].
# Display the percentages on the chart using autopct.
```

```python
import matplotlib.pyplot as plt

def plot_pie_chart():
    try:
        # Data
        labels = ['Python', 'Java', 'C++', 'JavaScript']
        sizes = [40, 25, 20, 15]
        colors = ['skyblue', 'lightgreen', 'gold', 'salmon']  # Different colors for clarity

        # Plot
        plt.figure(figsize=(8, 8))
        plt.pie(
            sizes,
            labels=labels,
            colors=colors,
            autopct='%1.1f%%',  # Display percentages with one decimal place
            startangle=140,     # Rotate the start angle for better visual balance
            shadow=True         # Add shadow for a 3D effect
        )

        # Customization
        plt.title('Programming Language Popularity', fontsize=16)
        plt.axis('equal')  # Ensure pie is drawn as a circle

        # Display the plot
        plt.show()

    except ImportError as e:
        print(f"Import error: {e}. Please ensure matplotlib is installed.")
    except Exception as e:
        print(f"An error occurred: {e}")

# Call the function
plot_pie_chart()
```
